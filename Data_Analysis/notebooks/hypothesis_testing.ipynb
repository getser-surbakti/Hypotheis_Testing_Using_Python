# 🧪 Kesimpulan Hypothesis Testing

- **Proporsi infeksi**: Kontrol = 40%, Perlakuan = 28%.  
- **Z-statistic**: 1.79  
- **p-value**: 0.073 (> 0.05)  
- **Confidence Interval**: Kontrol (30%–49%), Perlakuan (19%–37%).  

👉 Interpretasi: Perlakuan *cenderung* menurunkan infeksi, tetapi perbedaan ini **belum signifikan secara statistik** pada tingkat kepercayaan 95%.  


1. Import library & data processed

```python
import pandas as pd
import statsmodels.api as sm

control_clean = pd.read_csv("../data/processed/control_group_clean.csv")
treatment_clean = pd.read_csv("../data/processed/treatment_group_clean.csv")
```

2. Hitung proporsi infeksi

```python
infected_control = control_clean["Infected"].sum()
infected_treatment = treatment_clean["Infected"].sum()

n_control = len(control_clean)
n_treatment = len(treatment_clean)

print("Kontrol - Proporsi infeksi:", infected_control/n_control)
print("Perlakuan - Proporsi infeksi:", infected_treatment/n_treatment)
```

3. Uji dua proporsi (Z-test)

```python
count = [infected_control, infected_treatment]
nobs = [n_control, n_treatment]

stat, pval = sm.stats.proportions_ztest(count, nobs)
print("Z-statistic:", stat)
print("p-value:", pval)
```

4. Confidence Interval

```python
confint = sm.stats.proportion_confint(count, nobs, method='normal')
print("Confidence Interval:", confint)
```

Interpretasi Hasil
1. Perbedaan proporsi

    - Kontrol lebih tinggi (40%) dibanding perlakuan (28%).

    - Secara deskriptif, perlakuan tampak menurunkan infeksi.

2. Signifikansi statistik

    - p-value = 0.073 → lebih besar dari 0.05.

    - Artinya, perbedaan ini belum signifikan secara statistik pada tingkat kepercayaan 95%.

    - Dengan kata lain, kita tidak bisa memastikan perlakuan benar‑benar efektif; ada kemungkinan hasil ini terjadi karena variasi acak.

3. Confidence Interval

    - Interval proporsi kontrol: 30% – 49%.

    - Interval proporsi perlakuan: 19% – 37%.

    - Ada overlap antara kedua interval → mendukung kesimpulan bahwa perbedaan belum signifikan.

Intinya
1. Secara deskriptif: perlakuan menurunkan infeksi (40% → 28%).

2. Secara inferensial: hasil belum signifikan (p=0.073 > 0.05).

3. Kesimpulan: perlakuan cenderung efektif, tapi bukti statistik belum cukup kuat.
